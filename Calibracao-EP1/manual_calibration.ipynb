```python
import cv2
import matplotlib.pyplot as plt
```

```python
dX = 153
dY = 217
dZ = 261.5
```

```python
img = cv2.imread('images/image_book.jpg')
plt.imshow(img)
plt.show()
```

```python
xi = 640
xf = 2547
yi = 1015
yf = 3697

sub_img = img[yi:yf, xi:xf]
plt.imshow(sub_img)
plt.show()
```

```python
w = xf - xi
h = yf - yi

print('Largura, altura:', (w, h))
```

```python
fx = (w * dZ) / dX
fy = (h * dZ) / dY

print('Fx, Fy:', (fx, fy))
```

```python
print('Fx, Fy (millimeters):', ((504.56041667 * dZ) / dX, (709.6125 * dZ) / dY))
```
